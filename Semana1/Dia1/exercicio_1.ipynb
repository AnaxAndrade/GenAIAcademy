# Aplicando um algorítmo básico de aprendizagem
Nesta parte, veremos um algorítmo básico de aprendizagem dos parâmetros utilizando os dados de entrada

O algorítmo começa com um valor aleatório depois usa cada par de entrada e saída para ajustar o valor do parâmetro
Considere a equação de uma reta (basicamente indica que a saída **y** é a entrada **x** multiplicada por um número **m** . É justamente o valor desse **m** que o algorítmo tem que <i>"aprender"</i>

### Com os dados iniciais
Temos vários pares iniciais de dados de entrada e valores de saída esperados
<br>
(x_fornecido, y_esperado)<br> **ex**: (-4, -12)   (1,3)   (5, 15)

### Equação de uma reta
```
y_calculado = m * x_fornecido
```

### Cálculo de erro
Basicamente erro é a distância entre o esperado e o calculado
```
erro = y_esperado - y_calculado
```


### Correção de erro (conceito de backpropagação) - **Importante!!!!**
```
m = m + (velocidade_aprendizagem * erro * x_fornecido)
```


````python
# Inicializar m aleatoriamente
m = random.uniform(-1, 1)  # Valor aleatório entre -1 e 1
learning_rate = 0.01 # A taxa ou "velocidade de aprendizagem"
epochs = 10 # Fazer quantas iterações em cima dos meus dados

# Lista para armazenar os valores de m e erro a cada iteração
m_values = []
err_values = []

logs = []

print("Valor inicial de m:", m)

for epoch in range(epochs):
    for i in range(len(x)):
        y_calculated = m * x[i]
        error = y[i] - y_calculated
        #print(f"[{i}] Erro: {error}, y: {y[i]}, y_calc: {y_calculated}")
        m = m + learning_rate * error * x[i]
        logs.append(f"[{i}]  Erro: {error} ==> M atualizado: {m}")
        m_values.append(m) #Armazena o m para visualização posterior
        err_values.append(error) #Armazena o erro para visualização posterior

print("Valor final de m:", m)
````

````python
# (Opcional) Caso queira visualizar em texto como foram sendo atualizados os erros e os valores de m
for log in logs:
    print(log)
````

# Visualizar no gráfico a evolução dos valores

## Visualizar a evolução do valor do parâmetro procurado (m)
Dá para ver que ao longo do tempo o valor do m foi aproximando-se rapidamente no início e atualizando com valores cada vez menores há medida que foi aproximando do valor alvo (3). Indicando, que o algorítmo está convergindo e chegando a um valor "ótimo" do parâmetro procurado (m)

````python
# Mostrar em gráfico os valores de m
plt.figure(figsize=(8, 6))  # Ajustar o tamanho da figura (opcional)
plt.xlabel("Iterações")
plt.ylabel("Valor de m")
plt.title("Evolução do valor de m ao longo das iterações")
plt.plot(range(len(m_values)), m_values, marker='o', linestyle='-')  # 'o' para marcadores circulares, '-' para linha contínua
plt.grid(True)
plt.show()
````

## Visualizar a evolução do valor erro
Dá para ver que ao longo do tempo o valor do erro foi diminuindo e aproximando-se de 0. Indicando, também que o algorítmo está convergindo e chegando a um valor "ótimo" do parâmetro procurado (m)

````python
# Mostrar dos valores do erro
plt.figure(figsize=(8, 6))  # Ajustar o tamanho da figura (opcional)
plt.xlabel("Iterações")
plt.ylabel("Valor do Erro")
plt.title("Evolução do valor do erro ao longo das iterações")
plt.plot(range(len(err_values)), err_values, marker='o', linestyle='-', color='red')  # 'o' para marcadores circulares, '-' para linha contínua
plt.grid(True)
plt.show()
````
